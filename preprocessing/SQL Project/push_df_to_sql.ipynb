### Pushing dataframe as a whole to SQL server

https://www.dataquest.io/blog/sql-insert-tutorial/

```python
from sqlalchemy import create_engine
```

```python
# create sqlalchemy engine
engine = create_engine("mysql+pymysql://{user}:{pw}@34.67.199.229/{db}"
                       .format(user="root",
                               pw="odasodas",
                               db="recipes_database"))
```

```python
dataframe.to_sql('odas_data', con = engine, if_exists = 'append', chunksize = 1000)
```
