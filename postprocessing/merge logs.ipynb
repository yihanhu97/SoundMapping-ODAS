###### Combine Sensor CSV Files
    Group by Start Time:
    Ideally, for each 5 min gap, there are 4 recordings documented.

```python
from datetime import datetime

Cols=['sensor','File','End Time', 'Timestamp Length(ms)',
       'Total Timestamps', '0 source detected', '1 source detected',
       '2 source detected', '3 source detected', '4 source detected']
all_df=[]

for i in range(4):
    df=pd.read_csv('ODAS%d.csv'%i)
    df['sensor']=i
    df['start time']=[datetime.fromisoformat(t) for t in list(df['Start Time'])]
    df=df.set_index('start time')
    df=df.sort_index()
    df=df[Cols]
    all_df.append(df)
    
# all_df[0].head(3)
cdf=pd.concat(all_df)
cdf=cdf.sort_index()

"""
After concatenation, noticed that there are duplicated files generated by ODAS.
Need to clean out the duplicated rows
"""
boo=cdf.index.duplicated(keep='first')
cdf=cdf.loc[~boo]
# cdf.head(20)
cdf.to_csv('combined.csv')
```
